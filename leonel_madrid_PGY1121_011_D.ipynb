```python
def mostrar_menu():
    print("Bienvenido al cine seleccione una opción:")
    print("1. Comprar entradas")
    print("2. Mostrar ubicaciones disponibles")
    print("3. Ver listado de asistentes por ubicación")
    print("4. Salir")


def comprar_entradas(asientos_disponibles):
    cantidad = int(input("Ingresa la cantidad de entradas que deseas comprar (1-3): "))
    if cantidad < 1 or cantidad > 3:
        print("Cantidad inválida elija un número del 1 al 3.")
        return
    print("Asientos disponibles:")
    mostrar_asientos_disponibles(asientos_disponibles)
    seleccionar_asientos(asientos_disponibles, cantidad)


def mostrar_asientos_disponibles(asientos_disponibles):
    for i, fila in enumerate(asientos_disponibles):
        for j, asiento in enumerate(fila):
            numero_asiento = i * len(asientos_disponibles[0]) + j + 1
            if asiento == 0:
                categoria = determinar_categoria(numero_asiento)
                print(f"Asiento {numero_asiento}: {categoria} - Disponible")
            else:
                print(f"Asiento {numero_asiento}: x")


def determinar_categoria(numero_asiento):
    if numero_asiento <= 20:
        return "Categoría Platinum - 120.000"
    elif numero_asiento <= 50:
        return "Categoría gold - 80.000"
    else:
        return "Categoría silver - 50.000"


def seleccionar_asientos(asientos_disponibles, cantidad):
    asientos_seleccionados = []
    rut = input("Ingrese su RUT: ")
    for _ in range(cantidad):
        numero_asiento = int(input("Selecciona el número del asiento: "))
        if numero_asiento < 1 or numero_asiento > len(asientos_disponibles) * len(asientos_disponibles[0]):
            print("Asiento inválido elija un número de asiento válido.")
            return
        fila = (numero_asiento - 1) // len(asientos_disponibles[0])
        asiento = (numero_asiento - 1) % len(asientos_disponibles[0])
        if asientos_disponibles[fila][asiento] == 0:
            asientos_disponibles[fila][asiento] = 1
            asientos_seleccionados.append((numero_asiento, rut))
        else:
            print("El asiento seleccionado está ocupado elija otro.")
            return
    print("Has seleccionado los siguientes asientos:")
    for asiento in asientos_seleccionados:
        categoria = determinar_categoria(asiento[0])
        print(f"Asiento {asiento[0]}: {categoria} - RUT: {asiento[1]}")


def listar_asistentes(asientos_disponibles):
    print("Listado de asistentes por ubicación:")
    asistentes = []
    for i, fila in enumerate(asientos_disponibles):
        for j, asiento in enumerate(fila):
            numero_asiento = i * len(asientos_disponibles[0]) + j + 1
            if asiento == 1:
                rut = input(f"Ingrese el RUT del asistente para el Asiento {numero_asiento}: ")
                asistentes.append((numero_asiento, rut))
    asistentes_ordenados = sorted(asistentes, reverse=True)
    for asistente in asistentes_ordenados:
        categoria = determinar_categoria(asistente[0])
        print(f"Asiento {asistente[0]}: {categoria} - RUT: {asistente[1]}")


def programa_compra_entradas():
    asientos_disponibles = [[0] * 10 for _ in range(10)]
    while True:
        mostrar_menu()
        opcion = input("Selecciona una opción (1-4): ")
        if opcion == "1":
            comprar_entradas(asientos_disponibles)
        elif opcion == "2":
            mostrar_asientos_disponibles(asientos_disponibles)\\leonel_madrid_PGY1121_011_D.ipynb
        elif opcion == "3":
            listar_asistentes(asientos_disponibles)
        elif opcion == "4":
            print("adios")
            break
        else:
            print("Opción inválida. Por favor, elige una opción válida.")


programa_compra_entradas()
```
